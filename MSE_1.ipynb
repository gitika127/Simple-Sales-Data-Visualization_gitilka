```python
import matplotlib.pyplot as plt
import pandas as pd

# Get sales data from user
num_entries = int(input("Enter number of entries: "))
data = {'Product': [], 'Units Sold': [], 'Revenue': []}

for i in range(num_entries):
    product = input(f"Enter product name for entry {i+1}: ")
    units_sold = int(input(f"Enter units sold for {product}: "))
    revenue = float(input(f"Enter revenue for {product}: "))

    data['Product'].append(product)
    data['Units Sold'].append(units_sold)
    data['Revenue'].append(revenue)

# Convert data to DataFrame
df = pd.DataFrame(data)

# Create a bar chart
plt.figure(figsize=(8, 5))
plt.bar(df['Product'], df['Revenue'], color='skyblue')

# Add labels and title
plt.xlabel('Products')
plt.ylabel('Revenue')
plt.title('Sales Data Visualization')

# Show the plot
plt.show()

# Print sales data
print(df)
```
